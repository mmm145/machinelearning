## Self refrection mechanism

- React
  - one framework that has reasoning and acting 
  - structured cycle
    - reasoning > acting > observing.
**cycle of React**

- reasoning
  - just think what should be done
- acting
  - decide the action based on reasoning phase.
- observing
  - take an action and see what will happen.
- based on reasoning/acting/observing
  - decide the context.

**reflexion**

- self-improvement with feedback loop.
- dynamic memoery integration.
  - improve with integrating the information flexibily based on past experience and result
- binary rewarding system
  - good = good
  - bad = bad
- trajectory evaluation
  - based on the binary rewarding,
  - decide a next action.
- two shot learning
  - when you learn new thing
  - first learn a little
  - and then
  - use the experience to learn more. 
- error recovery mechanisms
  - when error happens.
  - learn how to deal with the error
  - and prepared for next learning.


**chain of hindsight**

- iterative improvement with explicit human feedback.
  - 1. generate multiple response
    2. human rank them
    3. convert it into natural language feedback
    4. fine tune with the sequence
