# Activation functions

activation function is used for capturing non-linearity of the output. 

**Type of Activation functions**

- Relu
  - Relu(x) = max(0,x)
  - f = lambda x: x*(x>0)
  - derivative = 1 (x>0) 0 (x<0)
  - Relu is not differentiable but you can set derivative of relu = 0 when [0,1]
  - +converge faster than sigmoid and tanh
  - +when active, behave like linear
  - +no saturation when x>0
  - -non negative > SGD zig zag
  - -is not derivable at x = 0, but it just can have 0 or 1, depending on program
  - -learning does not happen when Relu output the 0
    - to deal with this, you can use leaky relu
    - it works in same way as drop out.
- leaky Relu / PRelu activation 
  - Leaky Relu(x) = max( ax , x ) with small a
  - f = lambda x : x * (x>0) + a * x * (x<0)
  - +prevent model from learing when x<0
  - a is hyper parameter.
- exponential linear unit / (ELU)
  - max(a(exp(x)-1), x)
  - f = lambda x: x * (x>0) + a * (np.exp(x)-1) * (x<0)
  - +it can work even under 0
- sigmoid
  - $\sigma(x) = \frac{1}{(1 + e^{-x})}$
  - f = lambda x : 1.0 / (1.0 + np.exp(-x))
  - derivative = $\sigma (1 - \sigma)$
  - around 0, unit behaves linearly
  -  + it is differentiable everywhere
  -  -at extreme, the gradient become 0 > no learning from gradient
  -  sigmoid unit is not negative >> SGD do zig zag  
- softmax
  - $soft-max_i = \frac{\exp{linear_i}}{\sum_{j=1}^{c}\exp{linear_j}}$
  - p(y = i / x, parameter ) = soft_max_i)
  - Softmax Loss Function
    - $\frac{1}{m}\sum_{i=1}^{m}[log( \sum_{j=1}^{c} \exp{linear_j-with-x_i} ) - linear_{y_i}-with-x_i]$
- step
- tanh(hyperbolic tangent)
  - $tanh(x9 = 2\sigma(2x)-1$
  - f = lambda x: np.tanh(x)
  - +zero centered
  - +differentiable everywhere
  - +behaves linearly around zero
 
- softplus
  - log( 1 + exp(x))
  - similar to relu
  - +differentible at everywhere
  - usually works worse than relu
 
- swish(x)  = x * sigmoid(bx)

- GELU(x) = x * p(x<=X)
    

**properties of activation function**

- differentible.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sig = lambda x: 1/ (1 + np.exp(-x))
x = np.linspace(-10,10,100)
plt.plot(x,sig(x))
```

```python
x = np.linspace(-10,10,100)
plt.plot(x,np.tanh(x))
```

```python
relu = lambda x: x*(x>0)
x = np.linspace(-10,10,100)
plt.plot(x,relu(x))
```

```python
softplus = lambda x: np.log(1 + np.exp(x))
x = np.linspace(-10,10,100)
plt.plot(x, softplus(x))
```
