# confusion matrix
table where one axis is true label, and the other axis is prediction label.

TP = predict positve when true is positive

FP = predict positive when true is negative

TN = predict negative when true is negative

FN = predict negative when true is positive
# accuracy 
Accuracy = ( TP + TN ) / ( TP + TN + FP + FN )

**when**

- the dataset is balanced
-  general overview of model

# precision

precision  = TP / (TP + FP )

predicting positive when true is positive / (all the predicting positive)

**when**
- false positive are costly
- high precision > high recall 


# recall (sensitivity) 

recall = (true positive) / (true positive + false negative) 

recall = predicting positive / (all the positive)

- false negative are costly
- 

# f1 score 

mix of precision and recall 

f1 score = 2 * (precision * recall )/( precision + recall )

**when**

- dataset is imbalanced, and both false positive and false negative matter.

# specificity 

TN / (TN + FP)
